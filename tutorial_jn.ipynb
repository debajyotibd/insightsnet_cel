# Corpora and Corpus Information

To check all the existing indexed coropa, we can use **corpora_info()** function. \
In **corpora_info()** function we have to pass one parameter, which is registry path as string data type.

```python
from corporainfo import corpora_info

corpora_info("/home/test/Documents/cwb/registry")
```

To check any specific corpus's attributes(s-tag or p-tag) we can use **corpus_attributes()** function. \
To use **corpus_attributes()** function, we have to pass two parameters. First one is registry path, which is strig data type and the secone one is corpus name, which is also srting data type

```python
from corporainfo import corpus_attributes

corpus_attributes("/home/test/Documents/cwb/registry", "GP_DE")
```
